```python
#Cleanup the Salaries column first: 

#the following will edit the csv file DIRECTLY instead as a pandas dataframe
#search and replace the contents
text = open("output_.csv", "r") #opening file in read mode

#join() method combines all contents of output.csv and formed as a string
text = ''.join([i for i in text]) #issue with this method, we combine all contents into one string; not great when 
#trying to restructure the dataset, would need to use pandas instead

#replace unnecessary text in salaries column
text = text.replace("Employer est.:", " ") 
text = text.replace("(Glassdoor est.)", " ")
  
# output.csv is the output file opened in write mode
x = open("output_.csv","w") 
  
# all the replaced text is written in same output.csv file
x.writelines(text)
x.close()
#this should work
```

```python
#okay, we're going to focus only on Salaries
#reading this as a dataframe with pandas
df = pd.read_csv("output_.csv", encoding= 'unicode_escape')
df
```

```python
salaries = df['company_salary']
salaries = salaries.to_frame()
salaries = salaries['company_salary'].fillna(0) #ends up reverting back to series
salaries = salaries.to_frame()
salaries = salaries['company_salary'].astype(str) #keeps reverting back to series
salaries = salaries.to_frame()
salaries
```

```python
print(salaries)
```

```python
salaries = salaries.astype(str)

salaries['company_salary'] = salaries['company_salary'].fillna('0')

salaries["min_salary"] = None
salaries["max_salary"] = None

salaries.loc[salaries['company_salary'] == '0', ['min_salary', 'max_salary']] = 0
salaries['min_salary'] = salaries['company_salary'].str.extract(r'\$(\d{2,3})K - \$(\d{2,3})K')[0].fillna(0).astype(int)
salaries['max_salary'] = salaries['company_salary'].str.extract(r'\$(\d{2,3})K - \$(\d{2,3})K')[1].fillna(0).astype(int)

salaries['average_salary'] = salaries[['min_salary', 'max_salary']].mean(axis=1)
```

```python
print(salaries) #use average salary column for the summary statistics
```
