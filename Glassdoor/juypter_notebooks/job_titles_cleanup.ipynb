```python
df = pd.read_csv("output_.csv", encoding= 'unicode_escape')
df
```

```python
job_title = df['company_offeredRole']
job_title = job_title.to_frame()
job_title
```

```python
job_title["Role"] = None #create empty column 
job_title["Role"][job_title["company_offeredRole"].str.contains("Data Scientist", case=False)] = "Data Scientist" #fill the column with phrases we're looking for
job_title["Role"][job_title["company_offeredRole"].str.contains("Data Analytics", case=False)] = "Data Analyst" #fill the column
job_title["Role"][job_title["company_offeredRole"].str.contains("Applied Scientist", case=False)] = "Machine Learning"
job_title["Role"][job_title["company_offeredRole"].str.contains("Data Engineer", case=False)] = "Data Engineer" 
job_title["Role"][job_title["company_offeredRole"].str.contains("Software Engineer", case=False)] = "Others" 
job_title["Role"][job_title["company_offeredRole"].str.contains("Data Integration", case=False)] = "Others" 
job_title["Role"][job_title["company_offeredRole"].str.contains("Machine Learning", case=False)] = "Machine Learning" 
job_title["Role"][job_title["company_offeredRole"].str.contains("Systems Engineer", case=False)] = "Others" 
job_title["Role"][job_title["company_offeredRole"].str.contains("Business Analytics", case=False)] = "Data Analyst" 
job_title["Role"][job_title["company_offeredRole"].str.contains("Statistician", case=False)] = "Data Scientist" 
job_title["Role"][job_title["company_offeredRole"].str.contains("Data Science", case=False)] = "Data Scientist" 
job_title["Role"][job_title["company_offeredRole"].str.contains("Data Analyst", case=False)] = "Data Analyst" 

job_title
```
