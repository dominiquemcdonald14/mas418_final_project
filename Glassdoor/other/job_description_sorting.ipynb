```python
df = pd.read_csv("output_.csv", encoding= 'unicode_escape')
df #we have special characters issue here
```

```python
job_desc = df['listing_jobDesc']
job_desc = job_desc.to_frame()
job_desc
```

```python
job_desc = job_desc.listing_jobDesc.str.replace('\r\n', '') #remove newline
job_desc = job_desc.to_frame()
job_desc
```

```python
pd.set_option("max_colwidth", None)
job_desc
```

```python
job_desc
```

```python
job_desc["Bachelor_Degree"] = None #create empty column 
job_desc["Bachelor_Degree"][job_desc["listing_jobDesc"].str.contains("Bachelor's Degree", case=False)] = "Bachelor's Degree"

job_desc["Master_Degree"] = None  
job_desc["Master_Degree"][job_desc["listing_jobDesc"].str.contains("Master's Degree", case=False)] = "Master's Degree" 

job_desc["Ph_D"] = None 
job_desc["Ph_D"][job_desc["listing_jobDesc"].str.contains("Ph.D", case=False)] = "Ph.D" 

pd.reset_option("max_colwidth", None)
job_desc
```

```python
job_desc["R"] = None
job_desc["R"][job_desc["listing_jobDesc"].str.contains("R|\\bR programming\\b", case=False)] = "R" 

job_desc["SQL"] = None  
job_desc["SQL"][job_desc["listing_jobDesc"].str.contains("SQL", case=False)] = "SQL" 

job_desc["Python"] = None 
job_desc["Python"][job_desc["listing_jobDesc"].str.contains("Python", case=False)] = "Python"

job_desc
```
