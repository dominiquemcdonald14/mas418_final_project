```python
from linkedin_api import Linkedin
import re
import pandas as pd
import numpy as np

api = Linkedin('stats4182013@gmail.com', 'toolsindatascience')
```

```python
data_science = api.search_jobs(keywords = "Data Scientist", location_name = "Los Angeles")
```

```python
job_id = []

for x in range(len(data_science)):
    str = data_science[x]['dashEntityUrn']
    num = re.findall(r'\d+', str)
    job_id.append(num)
    
print(job_id)
```

```python
df = pd.DataFrame(columns = ['Job_Id', 'Company_Name', 'Job_Name', 'Location'], 
                  index = range(len(data_science)))

for index in range(10):
    job = api.get_job(int(job_id[index][0]))
    job_id_num = job_id[index][0]
    company_name = job['companyDetails']['com.linkedin.voyager.deco.jobs.web.shared.WebCompactJobPostingCompany']['companyResolutionResult']['name']
    job_name = job['title']
    location = job['formattedLocation']
    
    df.loc[index] = pd.Series({'Job_Id': job_id_num, 
                               'Company_Name': company_name, 
                               'Job_Name': job_name, 
                               'Location':location})

df
```
